```python
from functools import *

from numpy import *
from numpy.linalg import *
from numpy.random import *

from scipy.sparse.linalg import *
from scipy.optimize import *

def crand(*shape):
    return 1-2*rand(*shape) + 1j-2j*rand(*shape)

set_printoptions(linewidth=132)
```

```python
n = 3
nn = n*n
c = 2
cc = c*c
```

```python
def unitize(M):
    U, S, V = svd(M,full_matrices=False)
    return dot(U,V)

def symmetrize(M):
    return M + M.transpose().conj()

def is_unitary(M):
    return norm(identity(len(M.transpose().conj()))-dot(M.transpose().conj(),M))< 1e-14

def is_symmetric(M):
    return norm(M - M.transpose().conj())
```

```python
M = symmetrize(symmetrize(crand(nn,nn)).reshape(n,n,n,n).transpose(0,2,1,3).reshape(nn,nn))
print(norm(M-M.transpose().conj()))
```

```python
S, U = eigh(M)
UH = U.transpose().conj()
```

```python
print(norm(M-reduce(dot,[U,diag(S),UH])))
```

```python
T = reduce(dot,[U,diag(exp(-S)),UH])
```

```python
is_symmetric(T)
```

```python
X = unitize(symmetrize(crand(nn,nn)).reshape(n,n,n,n).transpose(0,2,1,3).reshape(nn,nn))

Xs = [X]
Zs = []

for _ in range(5):
    Z = dot(T,X)
    Zs.append(Z)
    Y = unitize(Z)
    print(norm(X-Y))
    X = Y
    Xs.append(X)
```
